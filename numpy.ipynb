```python
import numpy as np

a1 = np.zeros(4) # 创建一个由0组成的数组
print("a1={}".format(a1))

a2 = np.ones(5) # 创建一个由1组成的数组
print("a2={}".format(a2))

a3 = np.ones(5, dtype=np.float64)
print("a3:{}".format(a3))
```

```python
import numpy as np

a1 = np.empty(3)
print("a1:{}".format(a1))

a2 = np.empty(2)
print("a2:{}".format(a2))

a3 = np.empty(2, dtype=np.int64)
print("a3:{}".format(a3))
```

```python
import numpy as np

a1 = np.arange(4)
print("a1:{}".format(a1))

a2 = np.arange(4, 9, 2) # 第一个元素的值，最后一个元素的位置，步长
print("a2:{}".format(a2))
```

```python
import numpy as np

a1 = np.linspace(0, 10, num=5) # 生成一个从 0 到 10 之间均等间隔的 5 个元素
print("a1: {}".format(a1))

a2 = np.linspace(0, 10, num=5, endpoint=False) # 不包含结束值
print("a2: {}".format(a2))

# 返回样本间隔
result, step = np.linspace(0, 10, num=5, retstep=True)
print("result: {}".format(result))
print("step: {}".format(step))

# 指定类型
a3 = np.linspace(0, 10, num=5, dtype=float)
print("a3: {}".format(a3))


```

```python
# numpy 中的 cos() 和 sin() 函数
import numpy as np
import matplotlib.pyplot as plt

plt.title("Title")
plt.xlabel("X")
plt.ylabel("sin(x)")
a1 = np.linspace(-np.pi, np.pi * 2, num=256, endpoint=True)

cos_value = np.cos(a1)
sin_value = np.sin(a1)

# print(f"45 度（π/4 弧度）的余弦值：{cos_value}")
# print(f"45 度（π/4 弧度）的正弦值：{sin_value}")

plt.plot(a1, sin_value)

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
# 绘制散点图
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 轴数据
x = np.random.rand(100)
# 生成 y 轴数据
y = np.random.rand(100)
# 生成颜色数据
colors = np.random.rand(100)
# 生成大小数据
sizes = 1000 * np.random.rand(100)

# 绘制散点图
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
#Title
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
