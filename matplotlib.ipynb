```python
import matplotlib.pyplot as plt
import numpy as np

def f(t):
    return np.sin(t)

t = np.linspace(-2, 2, 100)

plt.plot(t, f(t))
plt.arrow(0, f(0), 0.5, f(0.5) - f(0), shape='full', lw=0, length_includes_head=True, head_width=.05)
plt.show()
```

```python
from pylab import *
import numpy as np

# figure(figsize=(10, 6), dip=80)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C, color='blue', label='cosine')
plt.plot(X, S, color='red', label='sine')
plt.grid(True) # 显示灰色网格

legend(loc='upper left', title='legend title') # 在左上角显示图例

plt.show()
```

```python
# 显式创建 figure 和 axes
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.rcParams['font.sans-serif'] = [
    'Noto Sans CJK SC',  # 思源黑体简体
    'WenQuanYi Micro Hei', 
    'WenQuanYi Zen Hei', # 文泉驿正黑
    'Microsoft YaHei'    # 微软雅黑（需手动安装）]  # Matplotlib中使用
]
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots()
ax.plot(x, x, label="线性")
ax.plot(x, x**2, label="二次函数")
ax.plot(x, x**3, label='立方')
ax.set_ylabel('y label')
ax.set_xlabel('x label')
ax.set_title('Simple Plot')
ax.legend()
plt.show()
```

```python
# 饼图
'''
函数原型：
matplotlib.pyplot.pie(
x, 
explode=None, 
labels=None, 
colors=None, 
autopct=None, 
pctdistance=0.6, 
shadow=False, 
labeldistance=1.1, 
startangle=0, 
radius=1, 
counterclock=True, 
wedgeprops=None,
textprops=None,
center=0,
0,
frame=False,
rotatelabels=False,
*, 
normalize=None,
data=None)
'''
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 190])

plt.pie(y, 
        labels=['A', 'B', 'C', 'D'], # 设置饼图标签
        colors=["#d5695d", "#5d8ca8", "#65a479", "#a564c9"], # 设置饼图颜色
        explode=(0, 0.2, 0, 0), # 第二部分突出显示，值越大，距离中心越远
        autopct='%.2f%%', # 格式化输出百分比
       )
plt.title('Title')
plt.show()
```

```python
# 散点图
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2

plt.scatter(x, y, s=area, c=colors, alpha=0.8)
plt.title("Title")
plt.show()
```

```python
# 绘制多图
import matplotlib.pyplot as plt
import numpy as np

# plot 1:
xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.subplot(1,2,1)
plt.plot(xpoints, ypoints)
plt.title('plot 1')

# plot 2:
x = np.array([1,2,3,4])
y = np.array([1,4,9,16])

plt.subplot(1,2,2)
plt.plot(x, y, linestyle='dotted', marker='o')
plt.title('plot 2')

plt.suptitle('title')
plt.show()
```
